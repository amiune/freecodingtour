### Copia de variables y referencias

Las variables que contienen un solo valor siempre se copian

```python
variable = 5
otra_variable = variable
print(variable)
print(otra_variable)
```

```python
otra_variable = 6
print(variable)
print(otra_variable)
```

Las variables que contienen mas de un valor solo se copia la referencia a menos que se explicite que se quiere hacer una copia

```python
lista = [1,2,3]
otra_lista = lista
print(lista)
print(otra_lista)
```

```python
otra_lista[1] = 500
print(lista)
print(otra_lista)
```

```python
lista = [1,2,3]
otra_lista = lista
print(lista)
print(otra_lista)
```

```python
otra_lista = [6,7,8]
print(lista)
print(otra_lista)
```

Debemos tener cuidado con esto cuando utilizamos funciones

```python
a = [1,2,3]
b = 2
print(b) #2
print(a) #[1,2,3]

def mifuncion(lista, variable):
  variable = 3
  print(variable) #3
  lista[2] = 5
  print(lista) #[1,2,5]

mifuncion(a,b)
print(b) #2
print(a) #[1,2,5]
```

```python
a = [1,2,3]
b = a
b.append(4)
print(a)
print(b)
```

### Alcance de las variable (Scope)

Si yo declaro una variable dentro de una funcion que tiene el mismo valor que una que esta fuera de la funcion seran dos variables distintas

```python
soy_global = 3
print(soy_global) #3

def mi_funcion():
  soy_local = 4
  print(soy_local) #4
  soy_global = 5 #mentira no es global, es local
  print(soy_global) #5

mi_funcion()

print(soy_global) #3
```

Si no la declaro puedo usarla pero no asignarle valores

```python
soy_global = 3
print(soy_global) #3

def mi_funcion():
  x = soy_global
  print(x)

mi_funcion()

print(soy_global) #3
```

```python
soy_global = 3
print(soy_global) #3

def mi_funcion():
  print(soy_global)
  soy_local = 4
  print(soy_local) #4
  soy_global = 5 #mentira no es global, es local
  print(soy_global) #5

mi_funcion()

print(soy_global) #3
```

```python
soy_global = 3
print(soy_global)

def mi_funcion():
  global soy_global
  soy_local = 4
  print(soy_local)
  soy_global = 5 #verdad! si es global
  print(soy_global)

mi_funcion()

print(soy_global)#5
```

```python
soy_global = 3
print(soy_global)

def mi_funcion():
  soy_local = 4
  print(soy_local)
  soy_global = 5 #mentira no es global, es local
  print(soy_global)
  return soy_global

soy_global = mi_funcion()

print(soy_global)
```
