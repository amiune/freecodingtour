# Environment: How to plot a board with matplotlib

```python
# prompt: create a matplotlib plot to show a 3x3 maze grid

import matplotlib.pyplot as plt
import numpy as np

def plot_maze_grid(maze):
  """
  Plots a 3x3 maze grid using matplotlib.

  Args:
    maze: A 2D list representing the maze.
            1 for wall, 0 for path.
  """
  fig, ax = plt.subplots()
  ax.set_xticks(np.arange(0, 3 + 1, 1))
  ax.set_yticks(np.arange(0, 3 + 1, 1))
  ax.grid(True, color='black')

  for i in range(3):
    for j in range(3):
      if maze[i][j] == 'X':
        ax.add_patch(plt.Rectangle((j, 2 - i), 1, 1, color='black'))

  plt.show()

# Example usage:
maze = [
    ['.', 'X', '.'],
    ['.', '.', '.'],
    ['.', '.', 'X']
]
plot_maze_grid(maze)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_tictactoe_board(board):
  """
  Plots a 3x3 maze grid using matplotlib.

  Args:
    board: A 2D list representing the maze.
            X for player1, O for player 2.
  """
  fig, ax = plt.subplots()
  ax.set_xticks(np.arange(0, 3 + 1, 1))
  ax.set_yticks(np.arange(0, 3 + 1, 1))
  ax.grid(True, color='black')

  for i in range(3):
    for j in range(3):
      if board[i][j] == 'X':
        ax.text(j+0.5, 2.5 - i, 'X', ha='center', va='center', fontsize=50, color='blue')
      elif board[i][j] == 'O':
        ax.text(j+0.5, 2.5 - i, 'O', ha='center', va='center', fontsize=50, color='red')

  plt.show()

# Example usage:
board = [
    ['.', 'X', '.'],
    ['.', '.', 'O'],
    ['O', '.', 'X']
]

plot_tictactoe_board(board)
```
