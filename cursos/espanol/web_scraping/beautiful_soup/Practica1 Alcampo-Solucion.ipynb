# Web scrapping de Supermercados Alcampo

Guardar la informacion (descripcion, precio y formato) de las cervezas en un dataframe de pandas

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://www.compraonline.alcampo.es/categories/bebidas/cervezas/cerveza-lata-est%C3%A1ndar/OC110701"
```

```python
response = requests.get(URL)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
productos = soup.find_all("div", attrs={"class":"product-card-container"})
print(len(productos))
```

```python
descripciones = []
precios = []
formatos = []
for p in productos:
    try:
        descripciones.append(p.find("h3").text.strip())
    except:
        descripciones.append(np.nan)

    try:
        formatos.append(p.find("div",attrs={"data-test":"fop-size"}).text.strip())
    except:
        formatos.append(np.nan)
    
    try:
        precios.append(p.find("span",attrs={"data-test":"fop-price"}).text.strip())
    except:
        precios.append(np.nan)
                  
df = pd.DataFrame({"Articulo":descripciones,"Precio":precios,"Formato":formatos})
df
```
