# ¬øQu√© es un test de hip√≥tesis?

Un **test de hip√≥tesis** es una forma **de tomar decisiones con datos**.
Sirve para responder preguntas del tipo:

> ‚Äú¬øParece que algo cambi√≥ o todo sigue igual?‚Äù



## Ejemplo simple

Imagina que una f√°brica produce chocolates que deber√≠an pesar **100 gramos**.
T√∫ sospechas que **la m√°quina est√° fallando** y que los chocolates pesan menos.

Entonces haces una prueba:

1. Tomas una muestra de chocolates.
2. Los pesas.
3. Usas la estad√≠stica para decidir si los resultados indican que **la m√°quina realmente cambi√≥** o si **solo fue casualidad**.


## La Receta Paso a paso

### 1. Planteamos dos hip√≥tesis

* **Hip√≥tesis nula (H‚ÇÄ)**: ‚ÄúNo pasa nada nuevo‚Äù.
  üëâ Los chocolates **pesan 100g como siempre**.

* **Hip√≥tesis alternativa (H‚ÇÅ)**: ‚ÄúS√≠ hay un cambio‚Äù.
  üëâ Los chocolates **no pesan 100g**.


### 2. Recolectamos datos

Pesas, por ejemplo, 30 chocolates y ves que pesan en promedio **98g**.


### 3. Calculamos la probabilidad (p-valor)

El **p-valor** te dice:

> ‚ÄúSi la m√°quina realmente funciona bien (H‚ÇÄ es cierta), ¬øqu√© tan probable es obtener un resultado tan raro como el tuyo?‚Äù

Si esa probabilidad es **muy baja**, significa que **tu resultado ser√≠a casi imposible si todo estuviera normal**.


### 4. Tomamos una decisi√≥n

* Si el **p-valor es peque√±o** (por ejemplo, menor que 0.05):
  üëâ Decimos: ‚ÄúEsto es demasiado raro, parece que la m√°quina cambi√≥‚Äù.
  ‚Üí **Rechazamos H‚ÇÄ**.

* Si el **p-valor es grande**:
  üëâ ‚ÄúEstos resultados pueden ser por casualidad.‚Äù
  ‚Üí **No rechazamos H‚ÇÄ**.


# Que tipo de preguntas podemos hacer?

**Todas las pruebas de hip√≥tesis comparan estimadores estad√≠sticos** (como medias, proporciones o varianzas), ya sea **entre s√≠** o **contra un valor poblacional te√≥rico**.

* **Estad√≠stico muestral:** valor calculado a partir de los datos de una muestra.
  Ejemplo: media muestral ( $\bar{x}$ ), proporci√≥n muestral ( $\hat{p}$ ), varianza muestral ( $s^2$ ).

* **Par√°metro poblacional:** valor verdadero (te√≥rico o conocido) de toda la poblaci√≥n.
  Ejemplo: media poblacional ( $\mu$ ), proporci√≥n poblacional ( $p$ ), varianza poblacional ( $\sigma^2$ ).


## Ejemplos:

### üîπ 1. **Muestra vs. Poblaci√≥n**

Comparas un **estad√≠stico muestral** con un **par√°metro poblacional** (que puede ser un valor hipot√©tico).

Ejemplo:
$$
H_0: \mu = 100
$$
$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

### üîπ 2. **Dos muestras independientes**

Comparas **dos estimadores muestrales** (por ejemplo, dos medias o proporciones).

Ejemplo:
$$
H_0: \mu_1 = \mu_2
$$
$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\text{error est√°ndar de la diferencia}}
$$

### üîπ 3. **Dos mediciones del mismo grupo (pareadas)**

Comparas **dos estimadores dependientes** (las medias antes y despu√©s).

$$
H_0: \mu_D = 0
$$

### üîπ 4. **M√°s de dos grupos (ANOVA)**

Comparas **varias medias muestrales** (estimadores)
para ver si podr√≠an venir de la **misma media poblacional**.

$$
H_0: \mu_1 = \mu_2 = \mu_3 = \dots
$$

### üîπ 5. **Chi-cuadrado**

Comparas **frecuencias observadas (muestrales)** con **frecuencias esperadas** (basadas en una hip√≥tesis poblacional o modelo te√≥rico).



## Ejemplo: Chocolates de 100 gramos

Supongamos que una f√°brica produce chocolates que **deber√≠an pesar 100 g**.
Queremos comprobar si la m√°quina est√° funcionando bien o no.


## 1. Planteamos las hip√≥tesis

* **Hip√≥tesis nula (H‚ÇÄ):**
  ( $\mu = 100$ )
  (el peso promedio de todos los chocolates es 100 g)

* **Hip√≥tesis alternativa (H‚ÇÅ):**
  ( $\mu \neq 100$ )
  (el peso promedio es distinto de 100 g)


## 2. Tomamos una muestra

Pesamos 30 chocolates y obtenemos:

* Media muestral: ( $\bar{x} = 98.5$ )
* Desviaci√≥n est√°ndar muestral: ( $s = 2.5$ )
* Tama√±o de la muestra: ( $n = 30$ )


##  3. Calculamos el estad√≠stico de prueba

Queremos medir **cu√°nto se aleja la media muestral (98.5)** del valor esperado (100), en unidades de error est√°ndar.

El **estad√≠stico t** se calcula as√≠:

$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

Sustituimos:

$$
t = \frac{98.5 - 100}{2.5 / \sqrt{30}} = \frac{-1.5}{0.456} \approx -3.29
$$


## 4. Buscamos el p-valor

Con ( n - 1 = 29 ) grados de libertad, consultamos una **tabla t de Student** o usamos Python/R/Excel.

El **p-valor ‚âà 0.0026** (en un test bilateral).


## 5. Tomamos la decisi√≥n

Elegimos un **nivel de significaci√≥n Œ± = 0.05**.

Como
$$
p\text{-valor} = 0.0026 < 0.05
$$

üëâ **Rechazamos H‚ÇÄ**.
Hay evidencia estad√≠stica de que el peso promedio **no es 100 g**.
Es muy probable que la m√°quina est√© fallando.


## Visualizacion

A continuacion graficamos la distribucion hipotetica del ejemplo del chocolate y la probabilidad de que la muestra pertenezca a esa distribucion.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Par√°metros para el ejemplo del chocolate con muestra peque√±a y sigma desconocida
mu = 100             # media poblacional hipotetizada
sample_mean = 98.5    # media muestral observada
s = 2.5                # desviaci√≥n est√°ndar muestral
n = 30               # tama√±o de la muestra (peque√±a -> usar distribuci√≥n t)
df = n - 1           # grados de libertad
alpha = 0.05         # nivel de significaci√≥n (para prueba bilateral)

# Calcular el estad√≠stico t
t_stat = (sample_mean - mu) / (s / np.sqrt(n))

# Calcular el p-valor (bilateral)
# Dado que t_stat es negativo, calculamos el √°rea de la cola izquierda y multiplicamos por 2
p_value = 2 * t.cdf(t_stat, df)

# Valores cr√≠ticos de t para alfa (bilateral)
t_crit_lower = t.ppf(alpha / 2, df)
t_crit_upper = t.ppf(1 - alpha / 2, df)

# Preparar los valores de la distribuci√≥n t
x = np.linspace(-4, 4, 500)
y = t.pdf(x, df)

plt.figure(figsize=(10, 6))

# Graficar la distribuci√≥n t
plt.plot(x, y, linewidth=2, color='blue', label='Distribuci√≥n t')

# Marcar la media cero (t = 0)
plt.axvline(0, linestyle='--', color='gray', label='Media (t=0)')

# Marcar el estad√≠stico t observado y su contraparte positiva
plt.axvline(t_stat, linestyle='--', color='red', label=f't-estad√≠stico = {t_stat:.2f}')
plt.axvline(-t_stat, linestyle='--', color='red') # |t_stat|

# Marcar los valores cr√≠ticos de t
plt.axvline(t_crit_lower, linestyle=':', color='green', label=f't cr√≠tico (Œ±/2) = {t_crit_lower:.2f}')
plt.axvline(t_crit_upper, linestyle=':', color='green', label=f't cr√≠tico (1-Œ±/2) = {t_crit_upper:.2f}')

# Sombrear el √°rea para el p-valor (dos colas)
# Sombrear la cola izquierda
plt.fill_between(x, y, where=(x <= t_stat), color='red', alpha=0.3, label=f'p-valor/2 = {p_value/2:.4f}')
# Sombrear la cola derecha
plt.fill_between(x, y, where=(x >= -t_stat), color='red', alpha=0.3)

# Anotar p-valor y alfa
plt.text(t_stat - 0.1, max(y) * 0.4, f"t = {t_stat:.2f}", fontsize=10, color='red', ha='right')
plt.text(-t_stat + 0.1, max(y) * 0.4, f"-t = {-t_stat:.2f}", fontsize=10, color='red', ha='left')

plt.text(t_crit_lower - 0.1, max(y) * 0.6, f"t_crit_inf = {t_crit_lower:.2f}", fontsize=10, color='green', ha='right')
plt.text(t_crit_upper + 0.1, max(y) * 0.6, f"t_crit_sup = {t_crit_upper:.2f}", fontsize=10, color='green', ha='left')


plt.text(-3.8, max(y) * 0.9, f"p-valor = {p_value:.4f}", fontsize=12, color='red')
plt.text(2.5, max(y) * 0.9, f"Œ± = {alpha}", fontsize=12, color='green')


plt.title(f"Distribuci√≥n t (gl = {df})\nMedia muestral = {sample_mean} g, Œº = {mu} g (Test Bilateral)", fontsize=14)
plt.xlabel("Valor t", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

## Resumen

| Estad√≠stico           | Cu√°ndo se usa                                                                                                                             | Tipo de variable                 | Ejemplo t√≠pico                                                                                     |
| --------------------- | ----------------------------------------------------------------------------------------------------------------------------------------- | -------------------------------- | -------------------------------------------------------------------------------------------------- |
| **z (normal)**        | Cuando **se conoce la desviaci√≥n est√°ndar de la poblaci√≥n (œÉ)** o cuando **n es grande (‚â• 30)** y la poblaci√≥n es aproximadamente normal. | Cuantitativa (num√©rica continua) | Comprobar si el peso promedio de una muestra (n=100) es diferente de 100 g, sabiendo que œÉ=2.      |
| **t (Student)**       | Cuando **NO se conoce œÉ**, solo tenemos la desviaci√≥n de la muestra (s), y/o la muestra es peque√±a (n < 30).                              | Cuantitativa                     | Ver si la presi√≥n arterial promedio en un grupo de 20 personas difiere del valor normal esperado.  |
| **œá¬≤ (chi-cuadrado)** | Para analizar **frecuencias o proporciones**, es decir, **variables categ√≥ricas**.                                                        | Categ√≥rica                       | Probar si la proporci√≥n de hombres y mujeres que prefieren un producto es diferente a lo esperado. |
| **F (Fisher)**        | Para **comparar varianzas** entre dos o m√°s grupos, o en **ANOVA** para comparar medias de varios grupos.                                 | Cuantitativa                     | Ver si las notas medias de tres colegios son iguales o no.                                         |


### Regla pr√°ctica
- Si comparas valores num√©ricos ‚Üí t o z.
- Si comparas categor√≠as ‚Üí œá¬≤.
- Si comparas m√°s de dos grupos ‚Üí F (ANOVA).
