# API Connection

1. La documentacion se puede ver aqui: https://www.last.fm/api
2. Obetener la API key desde aqui: https://www.last.fm/api/account/create
3. La direccion a la cual debe hacerle preguntas es: https://ws.audioscrobbler.com/2.0/ de acuerdo a que pregunta quiero hacer el final de esta url sera distinta, ver ejemplo a continuacion.


## Ejemplo de lectura de la documentacion y pregunta al servidor:

Como preguntarle al servidor cuales son las canciones mas escuchadas: https://www.last.fm/api/show/chart.getTopTracks

Me dice que tengo que preguntar a la direccion https://ws.audioscrobbler.com/2.0/?method=chart.gettoptracks&api_key=YOUR_API_KEY&format=json

y debo reemplazar el texto YOUR_API_KEY con la api key obtenida anteriormente, por ejemplo: https://ws.audioscrobbler.com/2.0/?method=chart.gettoptracks&api_key=e155e8d503ddd3b71b8feacbeaa9d2e6&format=json



```python
import requests

ret = requests.get(f"https://ws.audioscrobbler.com/2.0/?method=chart.gettoptracks&api_key=e155e8d503ddd3b71b8feacbeaa9d2e6&format=json")
ret.json()
```

## Como mostrar solo la informacion que yo quiero:

```python
for track in ret.json()["tracks"]["track"]:
  print(track["name"], ". Duracion:", track["duration"],"minutos")
```
