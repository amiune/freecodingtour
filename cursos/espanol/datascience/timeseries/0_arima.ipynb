# 🔹 1. ARIMA = AR + I + MA

Un modelo **ARIMA(p, d, q)** tiene tres partes:

* **p** → orden autoregresivo (AR)
* **d** → número de diferenciaciones (Integrated)
* **q** → orden de media móvil (MA)

---

# 🔹 2. Parte AR (AutoRegressive) — parámetro p

Un modelo **AR(p)** usa valores pasados de la serie.

## 🔸 AR(1)

$
y_t = \phi_1 y_{t-1} + \varepsilon_t
$

Depende solo del valor anterior.

---

## 🔸 AR(2)

$
y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \varepsilon_t
$

Aquí depende de los dos anteriores.

En general:

$
y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \varepsilon_t
$

👉 **p** = cuántos valores pasados usamos.

---

# 🔹 3. Parte I (Integrated) — parámetro d

Esto es **diferenciar la serie** para volverla estacionaria.

## 🔸 d = 1

$
y'_t = y_t - y_{t-1}
$

Restamos el valor anterior.

---

## 🔸 d = 2

$
y''_t = (y_t - y_{t-1}) - (y_{t-1} - y_{t-2})
$

O sea:

$
y''_t = y_t - 2y_{t-1} + y_{t-2}
$

👉 **d** = cuántas veces diferenciamos.

Ejemplo clásico:

Random walk:

$
y_t = y_{t-1} + \varepsilon_t
$

Si restamos:

$
y_t - y_{t-1} = \varepsilon_t
$

Se vuelve ruido blanco → por eso un random walk es ARIMA(0,1,0)

---

# 🔹 4. Parte MA (Moving Average) — parámetro q

La parte MA usa **errores pasados**, no valores pasados.

## 🔸 MA(1)

$
y_t = \varepsilon_t + \theta_1 \varepsilon_{t-1}
$

Depende del error anterior.

---

## 🔸 MA(2)

$
y_t = \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2}
$

En general:

$
y_t = \varepsilon_t + \theta_1 \varepsilon_{t-1} + \dots + \theta_q \varepsilon_{t-q}
$

👉 **q** = cuántos errores pasados usamos.

---

# 🔹 5. ARIMA completo

ARIMA aplica AR y MA **a la serie diferenciada**.

Ejemplo:

## 🔸 ARIMA(1,1,1)

Primero diferenciamos:

$
y'_t = y_t - y_{t-1}
$

Luego aplicamos ARMA(1,1):

$
y'_t = \phi_1 y'_{t-1} + \varepsilon_t + \theta_1 \varepsilon_{t-1}
$

Reemplazando:

$
(y_t - y_{t-1}) =
\phi_1 (y_{t-1} - y_{t-2})
$

* $ \varepsilon_t $
* $ \theta_1 \varepsilon_{t-1} $


Esa es la forma expandida real del modelo.

---

# 🔹 6. Intuición matemática simple

* **p** → memoria en niveles pasados
* **d** → elimina tendencia
* **q** → corrige errores recientes

Podemos verlo así:

$
\text{Nueva información} =
\text{memoria pasada}
+
\text{ruido actual}
+
\text{corrección por errores pasados}
$
---

# 🔹 7. Ejemplos rápidos

### ARIMA(0,0,0)

$
y_t = \varepsilon_t
$

Ruido blanco puro.

---

### ARIMA(1,0,0)

$
y_t = \phi_1 y_{t-1} + \varepsilon_t
$

Proceso autoregresivo simple.

---

### ARIMA(0,1,0)

$
y_t = y_{t-1} + \varepsilon_t
$

Random walk.

---

### ARIMA(2,1,0)

$
(y_t - y_{t-1}) =
\phi_1 (y_{t-1} - y_{t-2})
+
\phi_2 (y_{t-2} - y_{t-3})
+
\varepsilon_t
$
