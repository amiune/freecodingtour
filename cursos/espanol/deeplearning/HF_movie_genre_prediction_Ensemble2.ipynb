```python
import numpy as np
import pandas as pd
```

```python
df_xlnet = pd.read_csv('./xlnet_large_probs.csv', index_col="id")
df_roberta = pd.read_csv('./roberta_large_probs.csv', index_col="id")
df_albert_rotten = pd.read_csv('./albert_rotten_probs.csv', index_col="id")
df_xlnet_rotten = pd.read_csv('./xlnet_rotten_probs.csv', index_col="id")
df_multinomialnb = pd.read_csv('./multinomialnb_probs.csv', index_col="id")
```

```python
df_ensemble = ((df_xlnet**5) + (df_roberta**5)) / 2
df_ensemble.head(5)
```

```python
preds = []
for row in range(len(df_xlnet)):
    preds.append(df_xlnet.columns[np.argmax(df_ensemble.iloc[row,:])])
```

```python
preds[0:3]
```

```python
df_probs = pd.DataFrame({"id":df_xlnet.index,"genre":preds})
df_probs.to_csv("ensemble_submission2.csv", index=False)
```
