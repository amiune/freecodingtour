Substrings

```python
#Characters and substrings. Caracteres y subcadenas
frase = "Viva-la-vida"
print(len(frase))
print(frase[6]) #muestra el caracter que esta en la posicion 6 de la frase
print(frase[0]) #recordemos que el primer caracter esta en la posicion 0
```

```python
#Substrings [start:stop]
print(frase[2:8]) #muestra la substring que va de la posicion 2 a la 8-1
print(frase[:]) #si no pongo nada start es 0 y stop es el largo de la cadena
print(frase[0:12]) #en este caso 0 y 12
print(frase[:4]) #puedo omitir alguno de los dos
print(frase[3:]) #puedo omitir alguno de los dos
```

```python
#Que pasa si pongo numero por debajo de 0 o mayores al largo de la cadena?
print(frase[-1:100])
#Cosas raras! No lo usen de esta manera
```

```python
#Substrings [start:stop:step]
frase = "H-o-l-a-M-u-n-d-o"
print(len(frase))
print(frase[0:17:2])
print(frase[1:17:2])
```

```python
#Invertir el string
print(frase[::-1]) 
print(frase[::-2])
print(frase[6:1:-2])
```

```python
#Que pasa si pongo numero por debajo de 0 o mayores al largo de la cadena?
#Cosas raras! No lo usen de esta manera
print(frase[::-1]) 
print(frase[17:0:-1]) #Cuidado! No funciona igual que el anterior
print(frase[17:-1:-1]) #Cuidado! Este tampoco
```

```python
#Usando while o for para hacerlo manualmente
frase = "H-o-l-a-M-u-n-d-o"
print(frase)
largo = len(frase)
print(largo)

nueva_frase = ""
i = 0
while i < largo:
  nueva_frase += frase[i]
  i = i + 2
print(nueva_frase)
```

```python
nueva_frase = ""
for i in range(0,largo,2):
  nueva_frase += frase[i]
print(nueva_frase)
```

```python
nueva_frase = ""
i = 0
while i < largo:
  nueva_frase += frase[i]
  i = i + 2
print(nueva_frase)
```

```python
nueva_frase = ""
for i in range(0,largo,2):
  nueva_frase += frase[i]
print(nueva_frase)
```

Funcion range() y for

```python
#usando la funcion range con for
for i in range(5):
  print(i)
```

```python
#range(start,stop) genera los enteros desde start hasta stop (stop es omitido)
#En este caso desde 3 hasta 7 (porque 8 es omitido)
for i in range(3,8):
  print(i)
```

```python
#range(start,stop,step) 
#genera los enteros desde start hasta stop (stop es omitido) e incrementa usando step
#En este caso desde 4 hasta 12 (pero 12 es omitido) e incrementa de a 2
for i in range(4,12,2):
  print(i)
```

```python
#Puedo generar una secuencia descendente si step es negativo
#En este caso desde 12 hasta 4 (pero 4 es omitido) y decrementa de a 2
for i in range(12,4,-2):
  print(i)
```

```python
#Puedo utilizar la funcion range para generar listas de enteros
print(list(range(5)))
print(list(range(3,8)))
print(list(range(4,12,2)))
print(list(range(12,4,-2)))
```

```python
#for puede iterar sobre la funcion range, sobre listas, strings y otros elementos
for i in [20,5,7,100]:
  print(i)
```

```python
for i in "H-o-l-a-M-u-n-d-o":
  print(i)
```

```python
frase = "H-o-l-a-M-u-n-d-o"
indices = [8,0,14]
for i in indices:
  print(i)
  print(frase[i])
```
