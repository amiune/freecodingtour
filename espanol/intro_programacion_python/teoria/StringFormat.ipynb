## Como darle formato al texto que incluye variables

```python
name = 'Alicia'
age = 29

print(name, "tiene", age, "años")

print('%s tiene %d años' % (name, age))
print('{} tiene {} años'.format(name, age))
print(f'{name} tiene {age} años')

a = f'{name} tiene {age} años'
print(a)
```

## Multilinea

```python
name = 'Alicia'
age = 29
occupation = 'Cientifica de datos'

msg = (
    f'Nombre: {name}\n'
    f'Edad: {age}\n'
    f'Ocupacion: {occupation}'
)

print(msg)
```

## Fecha y hora

```python
import datetime

now = datetime.datetime.now()

print(f'{now:%Y-%m-%d %H:%M}')
print(f'{now:%A}')
print(f'{now:%a}')
print(f'{now:%B}')
print(f'{now:%j}')
print(f'{now:%w}')
```

## Numeros enteros y float

```python
val = 4.8
print(f'{val:10}')
print(f'{val:05}')
val2 = 100
print(f'{val2:05}')
print(f'{val:*<5}')
print(f'{val:.>5}')
print(f'{val:.^5}')
val1 = "Capitulo del libro"
val2 = 6
print(f'{val1:.<30}{val2:.>5}')
```

## Numeros con decimales (float)


```python
val = 1/7

#f'{value:{width}.{precision}}'
print(f'{val}')
print(f'{val:.2}')
print(f'{val:20.2%}')
print(f'{val:20.2}%')
```

## Tipos de notacion numerica

```python
val = 16

# hexadecimal en minusculas
print(f'{val:x}')
# hexadecimal en mayusculas
print(f'{val:X}')
# octal
print(f'{val:o}')
# binaria
print(f'{val:b}')
# cientifica
print(f'{val:e}')
```
