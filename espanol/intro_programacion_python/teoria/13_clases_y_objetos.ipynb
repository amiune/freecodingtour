# Clases y Objetos

Puedo encapsular varias variables dentro de una clase. Luego cada vez que cree un objeto de ese tipo de clase tendre todas esas varibles en mi objeto

```python
class Persona():
  nombre = ""
  edad = 0
```

```python
persona1 = Persona()
print("Nombre:",persona1.nombre)
print("Edad:",persona1.edad)
persona1.nombre = "Hernan"
persona1.edad = 39
print("Nombre:",persona1.nombre)
print("Edad:", persona1.edad)
```

```python
persona2 = Persona()
persona2.nombre = "Juan"
persona2.edad = 50
print(persona2.nombre)
```

Tambien puedo agregarle funciones

```python
class Persona():
  nombre = "sin nombre"
  edad = 0
  
  def hablar(self):
    print("Hola mi nombre es:",self.nombre)
```

```python
persona1 = Persona()
persona1.hablar()
```

Existe una funcion especial llamada \__init__(self) que se ejecuta solamente una vez para cada objeto de esa clase cuando es creado

> Indented block



```python
class Persona():
  nombre = "sin nombre"
  edad = 0

  def __init__(self, nombre, edad):
    self.nombre = nombre
    self.edad = edad
  
  def cambiarNombre(self, nombre):
    self.nombre = nombre

  def hablar(self):
    print("Hola mi nombre es:",self.nombre)
```

```python
persona2 = Persona("Hernan", 50)
persona2.hablar()
print(persona2.nombre)
print(persona2.edad)
```

Las clases se pueden extender heredando todas las variables y funciones de otra clase

```python
class Empleado(Persona):
  salario = 1
  tipo_de_moneda = "bitcoins"

  def pagar_salario(self):
    print("Se han enviado", self.salario, self.tipo_de_moneda)
```

```python
empleado1 = Empleado("Matilde")
empleado1.hablar()
empleado1.pagar_salario()
```

```python
class Empleado(Persona):
  salario = 1
  tipo_de_moneda = "bitcoins"

  def __init__(self, nombre, edad, salario):
    super().__init__(nombre, edad)
    self.salario = salario

  def pagar_salario(self):
    print("Se han enviado", self.salario, self.tipo_de_moneda)
```

```python
empleado1 = Empleado("Matilde", 39, 2)
empleado1.hablar()
empleado1.pagar_salario()
```
