Falacia del apostador

```python
n = 10 #numero de veces seguidas que quiero que salga negro
total_trials = 100000 #total de intentos a simular
n_black_in_row_counter = 0 #cantidad de veces que salio n veces seguidas negro
red_after_n_black_in_row_counter = 0 #cantidad de veces que salio rojo 
# numero de intentos
for trial in range(total_trials):
  counter = 0
  r = random.randrange(2) # 0 es negro 1 es rojo
  while r == 0 and counter < 5:
    r = random.randrange(2)
    counter += 1
  if counter == 5:
    n_black_in_row_counter += 1
    r = random.randrange(2)
    if r == 1:
      red_after_n_black_in_row_counter += 1

print(f"Total de intentos: {total_trials}")
print(f"Cantidad de veces que salieron {n} negros seguidos: {n_black_in_row_counter}")
prob_of_n_black = n_black_in_row_counter/total_trials
print(f"Probabilidad de {n} negros seguidos: {prob_of_n_black}")
print(f"Cantidad de veces que salio rojo luego de {n} negros seguidos: {red_after_n_black_in_row_counter}")
prob_of_red_after_n_black = red_after_n_black_in_row_counter/n_black_in_row_counter
print(f"Probabilidad rojo luego de {n} negros seguidos: {prob_of_red_after_n_black}")
```

Monty Hall. Las tres puertas

```python
total_trials = 10000 #total de intentos a simular
wins_changing_door = 0
wins_without_changing_door = 0
# numero de intentos
for trial in range(total_trials):
  doors = [0,0,0] #inicializamos las puertas con todas cabras (0 = cabra, 1 = auto)
  car_door = random.randrange(3) #numero de la puerta donde estara el auto
  #print(f"Puerta con auto: {car_door}")
  doors[car_door] = 1 #asignamos el auto a la puerta elegida aleatoriamente
  door_choosen_by_participant = random.randrange(3) #elijo una puerta al azar
  #print(f"Puerta elegida por el participante: {door_choosen_by_participant}")
  if door_choosen_by_participant == car_door:
    wins_without_changing_door += 1
  
  #guardo los numeros de las puertas que el conductor puede abrir
  goat_doors_numbers_to_open = [] 
  for i in range(3):
    #solo puede abrir donde no haya elegido el participante y donde haya una cabra
    if i != door_choosen_by_participant and doors[i] == 0:
      goat_doors_numbers_to_open.append(i)
  #print(f"Puertas que puede abrir el conductor: {goat_doors_numbers_to_open}")

  #elegir aleatoriamente la puerta a abrir
  door_to_open = random.randrange(len(goat_doors_numbers_to_open))
  door_to_change = [0,1,2]
  door_to_change.remove(door_choosen_by_participant) #eliminar la puerta elegida
  door_to_change.remove(goat_doors_numbers_to_open[door_to_open]) #eliminar la puerta abierta por el conductor
  #print(f"Puerta abierta por el conductor: {goat_doors_numbers_to_open[door_to_open]}")
  #print(f"Puerta a cambiar: {door_to_change}")
  if door_to_change[0] == car_door:
    wins_changing_door += 1
  #print()

print(f"Numero de veces que gano sin cambiar la eleccion: {wins_without_changing_door}")
print(f"Probabilidad: {wins_without_changing_door/total_trials}")
print(f"Numero de veces que gano cambiando la eleccion: {wins_changing_door}")
print(f"Probabilidad: {wins_changing_door/total_trials}")
```

100 prisioneros 50 casilleros

```python
#Estrategia aleatoria
number_of_prisoners = 100
total_trials = 10000 #total de intentos a simular
wins_counter = 0
# numero de intentos
for trial in range(total_trials):
  boxes = list(range(number_of_prisoners)) #genero los numeros en las cajas
  random.shuffle(boxes) #los mezclo aleatoriamente
  
  fail = False
  for prisoner in range(number_of_prisoners):
    #crear una lista de las posibles cajas a abrir
    boxes_to_choose = list(range(number_of_prisoners))
    
    tries = number_of_prisoners / 2
    #elegir una caja a abrir
    r = boxes_to_choose[random.randrange(len(boxes_to_choose))]
    while tries > 0 and boxes[r] != prisoner: 
      #abrir la caja y eliminarla de las posibles a elegir
      boxes_to_choose.remove(r)
      #elegir una caja a abrir
      r = boxes_to_choose[random.randrange(len(boxes_to_choose))]
      tries -= 1
    if tries == 0:
      fail = True
      break #si uno falla no hace falta seguir probando

  if not fail:
    wins_counter += 1

print(f"Numero de veces que se salvan: {wins_counter}")
print(f"Probabilidad: {wins_counter/total_trials}")
print(f"Probabilidad teorica: {0.5**number_of_prisoners}")
```
