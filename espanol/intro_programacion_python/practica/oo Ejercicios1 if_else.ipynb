### Ejercicios if else

Ejercicio 1:

Pedir que el usuario ingrese un numero. Mostrar luego si es positivo o negativo

*Ejemplos:*

*Ingrese un numero: 5*

*El numero ingresado es positivo*

*Ingrese un numero: -5*

*El numero ingresado es negativo*

Ejercicio 2:

Pedir al usuario que ingrese un password y mostrar si el password cumple con el requisito de contener al menos 6 caracteres.

*Ejemplos:*

*Ingrese un password: abcd12345*

*El password es aceptable*

*Ingrese un password: abcd*

*El password debe tener al menos 6 caracteres*


Ejercicio 3:

Pedir al usuario que ingrese un numero y decir si dicho numero es divisible por 3 y por 5 o no.

*Ejemplos:*

*Ingrese un numero: 3*

*El numero no es divisible por 3 y 5*

*Ingrese un numero: 15*

*El numero si es divisible por 3 y 5*

*Ingrese un numero: 18*

*El numero no es divisible por 3 y 5*

Ejercicio 4:

Pedir al usuario que ingrese dos numeros y que muestre como resultado su division. Si el segundo numero es cero que muestre la frase "No se puede dividir por cero"

*Ejemplos:*

*Ingrese el numerador: 5*

*Ingrese el divisor: 2*

*El resultado es 2.5*

*Ingrese el numerador: 5*

*Ingrese el divisor: 0*

*No se puede dividir por cero*

Ejercicio 5:

Pedir al usuario que ingrese un caracter y decir si dicho caracter es un digito entre 0 y 9

*Ejemplos:*

*Ingrese un caracter: A*

*El caracter introducido no es un digito*

*Ingrese un caracter: 5*

*El caracter introducido si es un digito*

*Ingrese un caracter: 30*

*El caracter introducido no es un digito*

Ejercicio 6:

Pedir al usuario que ingrese dos numeros y mostrar como resultado el mayor de los dos. Si son iguales mostrar cualquiera de ellos.

*Ejemplos:*

*Ingrese el primer numero: 5*

*Ingrese el segundo numero: 2*

*El mayor es 5*

*Ingrese el primer numero: 2*

*Ingrese el segundo numero: 2*

*El mayor es 2*

Ejercicio 7:

Pedir al usuario que ingrese un año y mostrar como resultado si es año bisiesto o no.
Un año es bisiesto si:
 - Si es divisible por 4 y no por 100
 - Si es divisible por 400

*Ejemplos:*

*Ingrese el año: 2000*

*El año 2000 si es bisiesto*

*Ingrese el año: 1900*

*El año 1900 no es bisiesto*

*Ingrese el año: 2004*

*El año 1900 si es bisiesto*

Ejercicio 8:

Pedir al usuario que ingrese un porcentaje de 0% a 100% y que de acuerdo a la siguiente tabla muestre el numero de estrellas que le corresponde:
 - Si porcentaje >  80% -> 5 estrellas
 - Si porcentaje >  60% -> 4 estrellas
 - Si porcentaje >  40% -> 3 estrellas
 - Si porcentaje >  20% -> 2 estrellas
 - Si porcentaje >=  0% -> 1 estrella

*Ejemplos:*

*Ingrese el porcentaje: 100%*

*5 estrellas*

*Ingrese el porcentaje: 21%*

*1 estrella*

Ejercicio 9:

Pedir al usuario que ingrese la cantidad de unidades electricas consumidas y que calcule el costo de la factura de electricidad dada la siguiente tabla:
 - Las primeras 50 unidades cuestan 0.50/unidad
 - Las siguientes 100 unidades cuestan 0.75/unidad
 - Las siguientes 100 unidades cuestan 1.25/unidad
 - Luego de 250 unidades cuestan 2.00/unidad

*Ejemplos:*

*Ingrese las unidades consumidas: 30*

*Costo total: 15 euros* (0.5 x 30)

*Ingrese las unidades consumidas: 100*

*Costo total: 62.5 euros* (0.5 x 50 + 0.75 x 50)

*Ingrese las unidades consumidas: 300*

*Costo total: 325 euros*
